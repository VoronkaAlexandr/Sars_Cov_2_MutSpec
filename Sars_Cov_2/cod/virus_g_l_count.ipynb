```python
import pandas as pd
```

```python
viruses_g_l_df = pd.read_csv('../data_obtain/viruses_g_l.csv', index_col=0)
```

```python
viruses_g_l_df
```

```python
aa_counter = {
    'Pro' : [],
    'Ser' : [],
    'Ala' : [],
    'Thr' : [],
    'Leu' : [],
    'Phe' : [],
    'Val' : [],
    'Ile' : [],
    'Met' : [],
    'Arg' : [],
    'Cys' : [],
    'Trp' : [],
    'Gly' : [],
    'His' : [],
    'Gln' : [],
    'Tyr' : [],
    'Asp' : [],
    'Glu' : [],
    'Asn' : [],
    'Lys' : []
}

amino_acids = list(aa_counter.keys())
```

```python
df = pd.DataFrame()
for strand in viruses_g_l_df['strand'].unique():
    strand_df = viruses_g_l_df[viruses_g_l_df['strand'] == strand]
    strand_gainers = []
    strand_loosers = []
    
    aa_counter = {
    'Pro' : [],
    'Ser' : [],
    'Ala' : [],
    'Thr' : [],
    'Leu' : [],
    'Phe' : [],
    'Val' : [],
    'Ile' : [],
    'Met' : [],
    'Arg' : [],
    'Cys' : [],
    'Trp' : [],
    'Gly' : [],
    'His' : [],
    'Gln' : [],
    'Tyr' : [],
    'Asp' : [],
    'Glu' : [],
    'Asn' : [],
    'Lys' : []
    }
    for aa in amino_acids:
        aa_counter[aa] = sum(strand_df['gainers'].str.contains(aa))
    aa_counter['strand'] = strand
    aa_counter['type'] = 'gainer'
    aa_counter['viruses_num'] = len(strand_df)
    
    aa_df = pd.DataFrame(aa_counter, index=[0])
    df = df.append(aa_df, ignore_index=True)
    
    aa_counter = {
    'Pro' : [],
    'Ser' : [],
    'Ala' : [],
    'Thr' : [],
    'Leu' : [],
    'Phe' : [],
    'Val' : [],
    'Ile' : [],
    'Met' : [],
    'Arg' : [],
    'Cys' : [],
    'Trp' : [],
    'Gly' : [],
    'His' : [],
    'Gln' : [],
    'Tyr' : [],
    'Asp' : [],
    'Glu' : [],
    'Asn' : [],
    'Lys' : []
    }
    for aa in amino_acids:
        aa_counter[aa] = sum(strand_df['loosers'].str.contains(aa))
    aa_counter['strand'] = strand
    aa_counter['type'] = 'looser'
    aa_counter['viruses_num'] = len(strand_df)
    
    aa_df = pd.DataFrame(aa_counter, index=[0])
    df = df.append(aa_df, ignore_index=True)
```
